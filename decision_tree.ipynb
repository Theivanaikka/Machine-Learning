```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
d = pd.read_csv("Book1.csv")
print(d.head())
print(d.tail())
```

```python
X= d[['Age','Study_Hours_per_Week','Sleep_Hours_per_Night','Attendance (%)']]
Y= d['Grade']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=7)
print(X_test)
print(Y_test)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,Y_train)
```

```python
p = clf.predict(X_test)
print("Predicted values:\n", p)
print("Original Values: ",Y_test)
```

```python
print("Classification Report:\n", classification_report(Y_test,p,zero_division=1))
print("Accuracy Score: ", accuracy_score(Y_test, p))
```

```python
print("Actual vs Predicted:\t")
comparison = pd.DataFrame({'Actual': Y_test.values, 'Predicted':p})
print(comparison.head(10))
```
