```python
df = np.array(df)
X = df[:,:-1]
Y = df[:, 1]
print(df)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
print(X_train[:4])
print(X_train.dtype)
Y_train_class = np.digitize(Y_train, bins=np.linspace(min(Y_train), max(Y_train), num=4))
print(Y_train_class.dtype)
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train,Y_train_class)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(np.concatenate((Y_train, Y_test)))
Y_train_encoded = le.transform(Y_train)
Y_test_encoded = le.transform(Y_test)
clf = KNeighborsClassifier()
clf.fit(X_train, Y_train_encoded)
```

```python
score = clf.score(X_train,Y_train_class)
sample_data_point = [[10, 15, 0, 0, 0, 9, 0, 8, 0, 6, 0, 2]]
print(clf.predict(sample_data_point))
```

```python
predict = clf.predict(X_test)
print(accuracy_score(Y_test_encoded, predict))
```
