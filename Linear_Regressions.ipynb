## ***Linear Regression***

# **Importing**

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **Data Allocating**

```python
X=[[15],[20],[22],[25], [28], [30], [32], [35], [38], [40], [42]]
Y=[25,30,35,41,55,65,68,71,75,80,99]
print(X)
print(Y)
plt.scatter(X, Y, color='Yellow', label='Actual Data')
plt.plot(X, Y, color='red', linestyle='-.', label='Predicted Line')
plt.xlabel("Token")
plt.ylabel("Fees")
plt.title("Token Vs Fees (Linear Regression)")
plt.legend()
```

# **Fit Data to the Model**

```python
r=LinearRegression()
r.fit(X,Y)
print(r.score(X,Y))
```

# **Prediction**

```python
print(r.predict([[38]]))
```

# **Plot**

```python
plt.scatter(X,Y,color="Green")
plt.plot(X,r.predict(X),color='pink')
plt.show()
```
